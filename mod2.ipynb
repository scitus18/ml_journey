```python
# whole bunch of train/test/validation data to be splitted now using the data.

n = len(df)
n_val = n_test = int(0.2*n)
n_train = n - (n_val + n_test)

idx = np.arange(n)
np.random.seed(2) # Used for fixing the random values.
np.random.shuffle(idx)

train_df = df.iloc[idx[:n_train]]
val_df = df.iloc[idx[n_train:n_val+n_train]]
test_df = df.iloc[idx[n_train+n_val:]]


train_df.head()
```

```python
if 'msrp' in train_df.columns : 

    train_df = train_df.reset_index(drop=True)
    test_df = test_df.reset_index(drop=True)
    val_df =val_df.reset_index(drop=True)

    y_train = np.log1p(train_df.msrp.values)
    y_test = np.log1p(test_df.msrp.values)
    y_val = np.log1p(val_df.msrp.values)

    del train_df['msrp']
    del test_df['msrp']
    del val_df['msrp']

else: 
    print("Already done")
    
    
sns.histplot(y_test, color='blue')
```
